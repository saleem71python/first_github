```python
import matplotlib.pyplot as plt
sns.boxplot(x="survived", 
            y="age",showmeans=True,
            meanprops={"marker":"+",
                      "markersize":"12",
                      "markeredgecolor":"red" },         
            data=kashti)
#show labels
plt.xlabel("How many survived",size=14),
plt.ylabel("age(years)",size=12),
plt.title("box plot of kitnay doobay or kitnay bach gaey",size=16,weight="bold")

plt.show()
```

# facet_wrap

```python
facet_wrap() creates a collection of plots (facets), where each
plot is differentiated by the faceting variable. 
These plots are wrapped into a certain number of columns or rows as specified 
by the user.
```

```python
facet_wrap() makes a long ribbon of panels (generated by any number 
of variables) and wraps it into 2d. This is useful if you have a single 
variable with many levels and want to arrange the plots in 
a more space efficient manner.

```

```python
#facet_grid
```

```python

facet_grid() forms a matrix of panels defined by row and column 
faceting variables. It is most useful when you have two discrete variables, 
and all combinations of the variables exist in the data. If you have 
only one variable with many levels, 
try facet_wrap() .
```

```python
FacetGrid class helps in visualizing distribution of one variable as well as 
the relationship between multiple variables separately within subsets 
of your dataset using multiple panels. A FacetGrid can be drawn with 
up to three dimensions ?
row, col, and hue.
```

```python
facet_grid(
  rows = NULL,
  cols = NULL,
  scales = "fixed",
  space = "fixed",
  shrink = TRUE,
  labeller = "label_value",
  as.table = TRUE,
  switch = NULL,
  drop = TRUE,
  margins = FALSE,
  facets = NULL
```
